# MatPlotLib
Library used for graphical representation of values/data.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
Data = {"Year" : [1920,1930,1940,1950,1960,1970,1980,1990,2000,2010,2020],
        "Exchange Rate" : [65,69,71,74,75,80,82,79,90,95,100]}

df = pd.DataFrame(Data, columns = ["Year", "Exchange Rate"])
```

```python
# To plot bar chart from the dataframe use .plot method of pandas and than see the chart using .show() method of matplotlib

df.plot(x = 'Year', y = 'Exchange Rate', kind = 'bar')
plt.show()
```

```python
df.plot(x = 'Exchange Rate', y = 'Year', kind = 'bar')
plt.show()
```

```python
df.plot(x = 'Year', y = 'Exchange Rate', kind = 'area')
plt.show()
```

```python
df.plot(x = 'Year', y = 'Exchange Rate', kind = 'line')
plt.show()
```

```python
df.plot(x = 'Year', y = 'Exchange Rate', kind = 'barh')  # horizontal chart
plt.show()
```

```python
df.plot(x = 'Year', y = 'Exchange Rate', kind = 'scatter')
plt.show()
```

```python
# Now using the matplotlib
plt.scatter(df['Year'], df['Exchange Rate'])
plt.show()
```

```python
plt.bar(df['Year'], df['Exchange Rate'])
plt.show()
```

```python
plt.barh(df['Year'], df['Exchange Rate'])
plt.show()
```

```python
Data = {"a" : [100,500,300]}
new_df = pd.DataFrame(Data, columns = ["a"], index = ['Pending', 'Complete','Ongoing'])
new_df.plot.pie(y = 'a',figsize = (5,5))
plt.show()
```
