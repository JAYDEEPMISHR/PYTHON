```python
import seaborn as sns
df = sns.load_dataset('iris') # load_dataset is one of the method in the seaborn and iris is predefined dataset given on the internet
print(df.head())
```

```python
print(len(df))
```

```python
# How many species we have?
count = df.species.value_counts()
print(count)
```

```python
count.plot(kind = 'bar')
```

```python
# kdeplot in seaborn (kernal density plot)
sns.kdeplot(df['petal_length'],fill = True) # instead of shade we will now use fill
```

```python
sns.kdeplot(df['petal_length'],fill = False) # instead of shade we will now use fill
```

```python
sns.kdeplot(df['petal_length']) # instead of shade we will now use fill
```
